```python
import numpy as np
import matplotlib.pyplot as plt

## define the sigmoid function
def sigmoid(z):
    return 1/(1+np.exp(-z))

# define points
x_points=np.array([-2,-1,1,2])
y_labels=np.array([0,0,1,1])


## define values for theta and theta0
parameters=[(-1,0),(-1,0),(0.5,0),(5,0)]

### Create a figure with 4 subplots
fig,axs=plt.subplots(2,2,figsize=(10,8))

for ax, (theta,theta0) in zip(axs.ravel(),parameters):
    # compute sigmoid function for range of x values
    x_range=np.linspace(-3,3,300)
    z=theta*x_range+theta0
    y=sigmoid(z)

    # plot the sigmoid curve
    ax.plot(x_range,y,label=f'Theta={theta},Theta0={theta0}')


    #### plot the points
    ax.scatter(x_points[:2], x_points[:2]-x_points[:2],color='r',label='Class -1')
    ax.scatter(x_points[2:], x_points[2:]-x_points[2:],color='blue',label='Class +1')

    ### Formatting the plot
    ax.axhline(y=0.5,color='gray', linestyle='--')
    ax.set_ylim(-0.1,1.1)
    ax.legend()

# Set overall figure title
fig.suptitle('Sigmois function for different theta and theta0')
plt.tight_layout(rect=[0,0.03,1,0.95])
plt.show()
```
